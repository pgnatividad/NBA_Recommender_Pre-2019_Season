##### I used try/except since some team urls lead to errors. This is due to some teams rebranding. 
##### ie: The Brooklyn Nets used to be the New Jersey Nets

```python
team_rosters=[]
year='2020.html'
for team in team_page_list[0:]:
    try:
        res = requests.get(url+team['url']+'2020.html')

        soup = BeautifulSoup(res.content, 'lxml')
        table = soup.find('table',{'id':'roster'}).find('tbody')
        for row in table.find_all('tr')[0:]:        
            player={}        
            player['name']=row.find('td').find('a').text
            player['pos'] =row.find_all('td')[1].text
#            print(row.find('td').find('a').text)
            player['team']=team['team_name']
            player['slug']=row.find('td').find('a').attrs['href']
            player['exp'] =row.find_all('td')[6].text
            team_rosters.append(player)
            time.sleep(3)
    except:
        #used to identify which teams lead to errors
        print("blame the"+team['team_name'])

players_df = pd.DataFrame(team_rosters)
```

### <u>Challenges with scrape</u>
I wanted to get the last 3 seasons of players however there are plenty of players with less than 3 years of experience. The first if statement is to catch if the players had 3+ years experience, 2 years, and 1 year. 

The try except in the loop is due to the code erroring out when looking at the team value of a player. If a player was traded mid season, the code would error. The try except catches and deals with this error.

The if statement in the loop is to put a value in "3pt percentage" if the player did not attempt any 3's. Without this, the value of the cell would be blank and not show up in isnull().

The outside try/except is to catch if there an error in the attempt at scraping. These errors could be due to trying to access the source page too much. The except shows which indicies errors occured.
